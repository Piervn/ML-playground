```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size = 0.2, random_state = 42, stratify = iris.target)

knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(X_train, y_train)
```

```python
y_predict = knn.predict(X_test)
print(f"Prediction for test data: {y_predict}")

score = knn.score(X_test, y_test) 
print(f"Accuracy: {score * 100:.2f}%") 
```

```python
n_neighbors = 50

train_accuracies = []
test_accuracies = []

for neighbors in range(1, n_neighbors + 1):
    knn = KNeighborsClassifier(neighbors)
    knn.fit(X_train, y_train)
    
    train_accuracies.append(knn.score(X_train, y_train))
    test_accuracies .append(knn.score(X_test, y_test))
```

```python
plt.plot(range(1, n_neighbors + 1), train_accuracies, label = "Train")
plt.plot(range(1, n_neighbors + 1), test_accuracies, label = "Test")
plt.xlabel("Number of neighbors")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```
